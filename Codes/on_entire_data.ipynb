# Random Forest

```python
# Test Train Split
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/6, random_state = 0)


#fitting Random forest regression model to the dataset
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
regressor.fit(X_train,y_train)


#Predicting a new result with Random forest regression 
y_pred = regressor.predict(X_test)

#RMS value
rms = sklearn.metrics.mean_squared_error(y_test, y_pred)**0.5
rms_mass = sklearn.metrics.mean_squared_error(y_test[:,0], y_pred[:,0])**0.5
rms_rad = sklearn.metrics.mean_squared_error(y_test[:,0], y_pred[:,0])**0.5
rms_lam = sklearn.metrics.mean_squared_error(y_test[:,0], y_pred[:,0])**0.5

#R2 value
r2 = sklearn.metrics.r2_score(y_test,y_pred)

result = result.append({'Model':'Random forest', 'RMS':rms,'RMS_NS_Mass':rms_mass,'RMS_NS_Rad':rms_rad,
                        'RMS_Lam':rms_lam, 'R2':r2}, ignore_index=True)
result
```

# Sampling from an MVGD and Testing

```python
#Sampling

#Input mu (mean) -> [e(0),rho(0),K(0),Q(0),J(0),L(0),Ksym(0)]
mu = [-16.0,0.16,230.0,300,32,60,-100]  #small errata in paper, the mean value of Q(0) = +300

#Input Sigma (covariance matrix)
#       [e,  rho,  K,  Q,  J,  L,  Ksym]
sigma =[[1,   0,   0,  0,  0,  0,    0], #e
        [0,   1,   0,  0,  0,  0,    0],#rho
        [0,   0,   1,  0,  0,  0,    0],#K
        [0,   0,   0,  1,  0,  0,    0],#Q
        [0,   0,   0,  0,  1,  0,    0],#J
        [0,   0,   0,  0,  0,  1,    0],#L
        [0,   0,   0,  0,  0,  0,    1]]#Ksym

number_of_samples = 100
X_test = np.random.multivariate_normal(mu,sigma,size=(number_of_samples,1,1))
#the rho(0) value in some of the samples is negative, we have to remove those
X_test = X_test[:,0,0]
X_new = []
i=0
for x in X_test:
    if x[1]>=0:
        X_new.append(x.tolist())
    i+=1
X_new = np.array(X_new)
print(X_new.shape)
X_test
```
